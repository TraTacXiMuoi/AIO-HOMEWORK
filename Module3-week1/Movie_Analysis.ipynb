```python
data.isnull().sum()
score_mean = data['Metascore'].mean()
data['Metascore'].fillna(score_mean, inplace = True)
```

```python
def rating_group(score):
    if score >= 75.0:
        return 'Good'
    elif score >= 55.0:
        return 'Average'
    else:
        return 'Bad'

data['Rating_by_score'] = data['Metascore'].apply(rating_group)
data[['Title','Director','Metascore','Rating_by_score']].head(3)
```
